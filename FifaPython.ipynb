```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print ('importé')
```

```python
df = pd.read_csv("data.csv")
```

```python
#graphique données continues avec la note des joueurs
x= df.Overall
sns.distplot(x)
```

```python
# Nous allons jeter un oeil aux joueurs les plus fort du jeu
mask2 = (df.Overall >= 90) 
sns.boxplot(x='Overall', y='Name', data=df[mask2])
```

```python
#digramme en camembert voulant montrer la valeure marchande des joueurs 
#(plus de 100 millions d'euros, entre 80 et 100 millions, entre 50 et 80 et enfin ceux de moins de 50)
plus_cent = np.sum(df.Value=="> €100M")
labels = '> 100 M', 'entre 80 et 100 M', 'entre 50 et 80 M', '<50 M'
sizes = [plus_cent, 80, 45, 40]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, 
        autopct='%1.1f%%', shadow=True, startangle=90)

plt.axis('equal')

plt.savefig('PieChart02.png')
plt.show()
```
