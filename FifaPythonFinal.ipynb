```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print ('importé')
```

```python
df = pd.read_csv("data.csv")
```

Le 1er graphique représente une gaussienne avec les notes des joueurs.
La note moyenne des joueurs est d'environ 66.
Un trou de données se situe entre 68 et 69 du a l'absence de joueurs.

```python
x= df.Overall
sns.distplot(x)
```

Le 2ème graphique représente les notes des joueurs en fonction de leur âge. 
Les notes sont assez homogènes entre 17 ans et 40 ans.
C'est assez exceptionnel que des joueurs de 17 ans soient presque aussi fort que des joueurs de 30 ans.
Il n'y pas de baton à 43 ans parce qu'il n'y pas de joueurs ayant 43 ans.

```python
fig = plt.figure()
names = df.Age
values = df.Overall
plt.bar(names, values);
plt.show()
```

Le 3ème diagramme est un diagramme en camembert qui montre la valeur marchande des joueurs.
L'écrasante majorité des joueurs a, soit une valeur marchande extrémement élevé (plus de 200 millions d'euros),
soit une valeur marchande assez faible (moins de 50 millions d'euros). 
Le reste des joueurs a une valeur marchande comprise entre 50 millions et 120 millions.

```python
df=df.replace('€','',regex=True)
df=df.replace('M','',regex=True)
plus_cent = np.sum(df.Value >= "200")
plus_quatrevingt = np.sum((df.Value >= "120") & (df.Value < "200"))
plus_cinquante = np.sum((df.Value >="50") & (df.Value <"80"))
moins_cinquante = np.sum(df.Value < "50")
labels = '> 200 M', 'entre 120 et 200 M', 'entre 50 et 80 M',  '<50 M'
sizes = [plus_cent, plus_quatrevingt, plus_cinquante,   moins_cinquante]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0.1, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, 
        autopct='%1.1f%%', shadow=True, startangle=90)

plt.axis('equal')

plt.savefig('PieChart02.png')
plt.show()
```
